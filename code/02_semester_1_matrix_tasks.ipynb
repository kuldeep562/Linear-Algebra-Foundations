```python
def prin(r, c, mat):
    print("Your matrix is:")
    for i in range(r):
        for j in range(c):
            print("\t", mat[i][j], end='')
        print()

def dia(mat, r):
    diagonal = []
    for i in range(r):
        diagonal.append(mat[i][i])
    return diagonal

def trac(dia):
    return sum(dia)

while True:
    r = int(input("Enter the number of rows: "))
    c = int(input("Enter the number of columns: "))
   
    if r != c:
        print("Rows and columns are not equal.")
        continue
    else:
        print("Enter the elements row-wise:")
        mat = []
        for i in range(r):
            row = []
            for j in range(c):
                ele = int(input("Enter the element: "))
                row.append(ele)
            mat.append(row)

        while True:
            print("\n\n-----------------------------------------------")
            print("Select the following options you want to do:\n")
            print("1-> Printing of matrices ")
            print("2-> Printing of diagonal ")
            print("3-> Finding trace ")
            print("4-> EXIT ")
            print("-----------------------------------------------\n")

            choice = int(input("\nEnter your choice: "))

            if choice == 1:
                prin(r, c, mat)
            elif choice == 2:
                print("\nDiagonal of your matrix is:", dia(mat, r))
            elif choice == 3:
                diagonal_elements = dia(mat, r)
                trace_value = trac(diagonal_elements)
                print("Trace of your matrix is:", trace_value)
            elif choice == 4:
                print("\nThank you.")
                break
            else:
                print("\nEnter a valid choice.")
           
            continue
        break
```
